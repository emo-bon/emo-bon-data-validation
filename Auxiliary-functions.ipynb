```python
import os
import csv
from pprint import pprint
from collections import Counter
```

```python
def process_sheet_type(sheet_names: list[str]) -> list:
    print(f"Total of {len(sheet_names)} sheets")
    
    #sheet_names = ["HCMR-1_water_column_sampling_validated.csv"]
    
    fields = []
    for sheet_name in sheet_names:
        print(f"Processing {sheet_name}")
        sheet_dict = csv.DictReader(open(os.path.join("./logsheets", sheet_name)))
    
        # Sheet_dict is a list(rows: dict)
        # Just take keys from first row
        for row in sheet_dict:
            pprint(f"Found {len(row.keys())} in sheet {sheet_name}")
            fields.append(list(row.keys()))
            break
        

    assert (len(fields) == len(sheet_names))
    return fields

sheet_types = [
[sheet for sheet in os.listdir("./logsheets") if sheet.split("_", 1)[1] == "water_column_sampling_validated.csv"],
[sheet for sheet in os.listdir("./logsheets") if sheet.split("_", 1)[1] == "soft_sediment_sampling_validated.csv"],
[sheet for sheet in os.listdir("./logsheets") if sheet.split("_", 1)[1] == "water_column_measured_validated.csv"],
[sheet for sheet in os.listdir("./logsheets") if sheet.split("_", 1)[1] == "soft_sediment_measured_validated.csv"]
    ]
for sheet_names in sheet_types:
    fields = process_sheet_type(sheet_names)
    flat_fields = [field for sheet in fields for field in row]
    #pprint(f"Fields = {flat_fields}")
    counter = Counter(flat_fields)
    pprint(counter)
```
